```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
import shutil
```

```python
DIRTY_DATA_PATH = 'data_dirty/train/images/'
DIRTY_LAND_PATH = 'data_dirty/train/landmarks.csv'

DIRTY_IMAGE_PATHS = np.array([os.path.join(DIRTY_DATA_PATH , f) for f in os.listdir(DIRTY_DATA_PATH )])
DIRTY_LANDMARKS = pd.read_csv(DIRTY_LAND_PATH, sep='\t').set_index('file_name')
```

```python
PATH_TO_CLEAN_DATA = 'data_clean'
```

```python
noisy = pd.read_csv('noisy_images.csv')['file'].tolist()
```

```python
shit = pd.read_csv('./my_shit_images.csv',index_col=0).iloc[:,0].tolist()
```

```python
path_to_clean_images = [path for path in DIRTY_IMAGE_PATHS if os.path.split(path)[-1] not in noisy and os.path.split(path)[-1] not in shit]
```

```python
len(path_to_clean_images),len(DIRTY_IMAGE_PATHS),len(shit),len(noisy)
```

```python
os.makedirs(os.path.join(PATH_TO_CLEAN_DATA,'images'), exist_ok=True)
```

```python
for path in tqdm(path_to_clean_images): 
    shutil.copy(path, os.path.join(PATH_TO_CLEAN_DATA,'images',os.path.split(path)[-1]))
```

```python
DIRTY_LANDMARKS
```

```python
clean_landmarks = DIRTY_LANDMARKS[DIRTY_LANDMARKS.index.map(lambda x: x not in noisy and x not in shit)]
```

```python
assert len(path_to_clean_images) == clean_landmarks.shape[0]
```

```python
clean_landmarks.to_csv(os.path.join(PATH_TO_CLEAN_DATA,'landmarks.csv'),sep='\t')
```
